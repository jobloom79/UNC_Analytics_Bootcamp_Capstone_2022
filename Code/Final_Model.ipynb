## Preprocessing data for supervised learning

```python
#Drop the unnecessary columns 
clean_animal_data_df= Animal_Data_df.drop(['animal_id', 'state', 'sex', 'animal_type', 'breed_class', 'color'], axis=1)
clean_animal_data_df.head()
```

```python
#Check the status of column ['is_4dx_tasted'] 
clean_animal_data_df['is_4dx_tested'].value_counts()
```

```python
test_df = clean_animal_data_df.loc[clean_animal_data_df['is_4dx_tested'] != 'Not Tested']
test_df
```

```python
test_df['is_4dx_tested'].value_counts()
```

```python
#Converting the age column to just numbers - years
# First, splitting the column into two new colummns - one for number and the other for the string (years, months, etc.)
test_df[['age_num', 'age_str']] = test_df['age'].apply(lambda x: pd.Series(str(x).split(" ")))

# Setting the original age column equal to the age_num column - i.e. changing the data in the age column to just the numbers
test_df['age'] = test_df['age_num']

# Changing the data type of the age column to float
test_df['age'] = test_df['age'].astype('float')
```

```python
# Checking the column to see what the unique string values are
test_df["age_str"].unique()
```

```python
# Using the string column to change the number value - converting all ages to years
test_df.loc[test_df.age_str == 'days', ['age']] = test_df['age'] / 365
test_df.loc[test_df.age_str == 'day', ['age']] = test_df['age'] / 365
test_df.loc[test_df.age_str == 'months', ['age']] = test_df['age'] / 12
test_df.loc[test_df.age_str == 'month', ['age']] = test_df['age'] / 12
test_df.loc[test_df.age_str == 'weeks', ['age']] = test_df['age'] / 52
test_df.loc[test_df.age_str == 'week', ['age']] = test_df['age'] / 52

print(test_df.shape)
test_df.head()
```

```python
# Dropping the extra (created) columns - age_num and age_str
prep_animalData_df = test_df.drop(['age_str', 'age_num'], axis=1)
print(prep_animalData_df.columns)
print(prep_animalData_df.dtypes)
print(prep_animalData_df.shape)
prep_animalData_df.head()
```

```python
# Creating features
X = test_df.drop(columns='is_4dx_tested')
X = pd.get_dummies(X)

#Find A Target
y = test_df['is_4dx_tested']

X.head(2)
```

```python
X.describe()
```

```python
y.value_counts()
```

## Split our data into training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Random Oversampling

```python
ros = RandomOverSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data

lr_model = LogisticRegression(solver='lbfgs', random_state=42)
lr_model.fit(X_resampled, y_resampled)
```

```python
y_pred = lr_model.predict(X_test)
print(str('Random Oversampling'))
print(str('Balanced Accuracy Score'))
print(balanced_accuracy_score(y_test, y_pred))
```

```python
# Display the confusion matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(confusion_matrix, index=["Actual Negative", "Actual Positive"], columns=["Predicted Negative", "Predicted Positive"])
cm_df
```

```python
# Print the imbalanced classification report
#print(str('Random Oversampling'))
#print(str('Imbalanced Classification Report'))
#print(classification_report_imbalanced(y_test, y_pred))
report_df = pd.DataFrame(classification_report_imbalanced(y_test, y_pred, output_dict=True)).transpose()
report_df
```

## Random Forest

```python
df2 = test_df.drop(['age_str', 'age_num'], axis=1)

df2.head()
```

```python
def change_string(is_4dx_tested):
    if is_4dx_tested == 'Positive':
        return 1
    else:
        return 0

df2['is_4dx_tested'] = df2['is_4dx_tested'].apply(change_string) 
df2.head(5)             
```

```python
# Creating features set
X = df2.drop(columns='is_4dx_tested')
X = pd.get_dummies(X)
X.head()
```

```python
#Define the Target Set
y = df2['is_4dx_tested'].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
predictions
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual Negative", "Actual Positive"],columns=["Predicted Negative", "Predicted Positive"])

cm_df
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```
